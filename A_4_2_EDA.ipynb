## Part 2: Visualizing a Mel Spectrogram

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
song = '/content/pop450.wav'
```

```python
y,sr = librosa.load(song,duration=3)
```

```python
print(len(y))
```

```python
print(sr)
```

```python
mels = librosa.feature.melspectrogram(y=y,sr=sr)
```

```python
mels.shape
```

```python
mels_to_db = librosa.power_to_db(mels,ref=np.max)
```

```python
mels_to_db.shape
```

```python
plt.imshow(mels_to_db)
```

```python
fig, ax = plt.subplots()
img = librosa.display.specshow(mels_to_db, x_axis='time',
                         y_axis='mel', sr=sr,
                         ax=ax, cmap='viridis')
fig.colorbar(img, ax=ax, format='%+2.0f dB')
ax.set(title='3-Second Mel-frequency spectrogram for "Lady Marmalade"')
ax.set_ylim(8200, 0)
plt.show()
fig.savefig('/content/lady_marmalade.png')
```
